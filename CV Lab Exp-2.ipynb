```python
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread(r"C:/Users/Ganpati Kumar Roy/File/Pictures/Personal Data/Photo/Ganpati Roy.png")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
dimensions = img1.shape
print('Image Dimension    : ',dimensions)
```

```python
img1 = cv2.resize(img1, (400, 400))
```

##### Split the component of image and display BGR of images.

```python
B1, G1, R1 = cv2.split(img1)
cv2.imshow("B1", B1)
cv2.imshow("G1", G1)
cv2.imshow("R1", R1)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img2 = cv2.imread(r"C:/Users/Ganpati Kumar Roy/File\Pictures/Personal/Test/Amit Shah.jpg")
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
dimensions = img2.shape
print('Image Dimension    : ',dimensions)
```

```python
img2 = cv2.resize(img2, (400, 400))
```

##### Split the component of image and display BGR of images.

```python
B2, G2, R2 = cv2.split(img2)
cv2.imshow("B2", B2)
cv2.imshow("G2", G2)
cv2.imshow("R2", R2)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### Blend the images

```python
Blend_img = cv2.addWeighted(img1, 0.3, img2, 0.7, 0.0);
```

```python
cv2.imshow('Blend-Img', Blend_img)
 
cv2.waitKey(0)
cv2.destroyAllWindows()
```
