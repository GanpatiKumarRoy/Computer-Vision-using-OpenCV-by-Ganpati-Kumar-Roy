### Task-1

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
face_cascade = cv2.CascadeClassifier((cv2.data.haarcascades+'haarcascade_frontalface_default.xml'))
webcam = cv2.VideoCapture(0)
```

```python
while True:
    frame_read,frame = webcam.read()
    greyscale_frame = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    face_cordinates = face_cascade.detectMultiScale(greyscale_frame)
    for (x,y,w,h) in face_cordinates:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),5)

    cv2.imshow("Ganpati Roy",frame)
    cv2.waitKey(1) 
```

```python
img = cv2.imread(r"C:/Users/Ganpati Kumar Roy/File/Pictures/Personal Data/Photo/Ganpati Roy-2.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

##### Image Blurring: -

###### Average Bluring: -

```python
average_blur = cv2.blur(img, (20, 20))
plt.imshow(average_blur)
```

###### Gaussian Bluring: -

```python
gaussian_blur = cv2.GaussianBlur(img, (5,5), 0)
plt.imshow(gaussian_blur)
```

###### Median Bluring: -

```python
median_blur = cv2.medianBlur(img, 5)
plt.imshow(median_blur)
```
