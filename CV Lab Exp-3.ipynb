```python
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
face_cascade = cv2.CascadeClassifier((cv2.data.haarcascades+'haarcascade_frontalface_default.xml'))
```

#### Read the image

```python
img = cv2.imread(r"C:/Users/Ganpati Kumar Roy/File/Pictures/Personal Data/Photo/Ganpati Roy.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
```

```python
faces = face_cascade.detectMultiScale(img, scaleFactor = 1.20, minNeighbors = 4)
for x, y, w, h in faces:
    img = cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 250), 5)
    print("Face Detected Successfully.....!")
    print("Found {0} faces in this picture....!". format(len(faces)))
    plt.imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Reading Group Image

```python
group_img = cv2.imread(r"C:/Users/Ganpati Kumar Roy/File/Pictures/Camera Roll/Test/Ganpati.jpg")
group_img = cv2.cvtColor(group_img, cv2.COLOR_BGR2RGB)
plt.imshow(group_img)
```

```python
gray_image = cv2.cvtColor(group_img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
```

```python
faces = face_cascade.detectMultiScale(group_img, scaleFactor = 1.3, minNeighbors = 4)
for (x, y, w, h) in faces:
    img2 = cv2.rectangle(group_img, (x, y), (x+w, y+h), (255, 0, 0), 10)
    print("Face Detected Successfully.....!")
    print("Found {0} faces in this picture....!". format(len(faces)))
    plt.imshow(img2)
    break
cv2.waitKey(0)
cv2.destroyAllWindows()
```
