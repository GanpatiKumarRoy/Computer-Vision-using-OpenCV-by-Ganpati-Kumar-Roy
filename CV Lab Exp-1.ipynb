```python
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"C:/Users/Ganpati Kumar Roy/File/Pictures/Personal Data/Photo/Ganpati Roy.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# Check dimension :-
dimensions = img.shape
print('Image Dimension    : ',dimensions)
```

```python
# Gray scale image :-
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img1, cmap='gray')
```

```python
# Check dimension :-
dimensions = img1.shape
print('Image Dimension    : ',dimensions)
```

```python
scale_percent = 20 # percent of original size
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
dim = (width, height)
  
# resize image
resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
 
print('Resized Dimensions : ',resized.shape)
```

```python
# Crop the image :-
img = cv2.imread(r"C:/Users/Ganpati Kumar Roy/File/Pictures/Personal Data/Photo/Ganpati Roy.png")
roi = cv2.selectROI(img)
img
print(roi)
img_cropped = img[int(roi[1]):int(roi[1]+roi[3]),
            int(roi[0]):int(roi[0]+roi[2])]
cv2.imshow("Cropped Image", img_cropped)
cv2.waitKey(10000)        #Visible timing of cropped image
cv2.destroyAllWindows()
```
